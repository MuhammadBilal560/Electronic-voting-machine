```python
import numpy as np

# 1D Array
array_1d = np.arange(10)
print("1D Array:", array_1d)

# 2D Array
array_2d = np.arange(1, 10).reshape(3, 3)
print("2D Array:\n", array_2d)

# 3D Array
array_3d = np.random.randint(1, 101, size=(2, 3, 4))
print("3D Array:\n", array_3d)
```

```python
import numpy as np

# Assuming arr2 and arr3 are defined as follows:
arr2 = np.arange(1, 10).reshape(3, 3)
arr3 = np.random.randint(1, 101, size=(2, 3, 4))

# 1.1.1 Extract the second row from arr2
second_row = arr2[1]
print("Second Row:", second_row)

# 1.2.1 Extract the element at the 3rd row and 2nd column of arr2
element = arr2[2, 1]
print("Element at 3rd row, 2nd column:", element)

# 1.3.1 Extract all elements from the 3D array arr3 at the first index (0th layer)
first_layer = arr3[0]
print("First Layer of 3D Array:\n", first_layer)
```

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the line plot
plt.plot(x, y, marker='o')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Basic Line Plot')

# Adding grid lines
plt.grid()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create the scatter plot
plt.scatter(x, y, s=100, color='blue', alpha=0.6, edgecolors='w')

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot')

# Show the plot
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Categories and values
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Customize colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')

# Show the plot
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')

# Show the plot
plt.grid(axis='y')
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]

# Explode the largest section (Python)
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (Python)

# Create the pie chart
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie chart is a circle
plt.axis('equal')

# Title for the chart
plt.title('Programming Language Popularity')

# Show the plot
plt.show()

```
